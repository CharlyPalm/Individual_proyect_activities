Identify common tokens and resolve

the first approach, delete all, is not right; delete those elements that are less frequent. Example: paper [105   0   5] delete for segment 15 and 32, if the number is the same, delete the word is all bags of words.

*There* are 201 terms.

Create three lists to keep track of which tokens will be erased for each segment.


```python
i = 0
commonTerms = []
elementsDeleted = []
segment14delete = []
segment15delete = []
segment32delete = []
for (columnName, columnData) in df.items():
    nOcurrences = sum(x != 0 for x in columnData.values)
    if nOcurrences > 1:
      i = i + 1
      commonTerms.append(columnName) # keep track of all the common terms
      print('Token : ', columnName)
      print('Token Frequency : ', columnData.values)
      print('Token Repetition', sum(x != 0 for x in columnData.values))
      segment14count = columnData.values[0]
      segment15count = columnData.values[1]
      segment32count = columnData.values[2]
      print('Segment 14 Token count ', segment14count)
      print('Segment 15 Token count ', segment15count)
      print('Segment 32 Token count ', segment32count)
      # Less frequent tokens are deleted from the bag of words
      if (segment14count > segment15count) and (segment14count > segment32count):
        print('Segment 14 wins')
        segment15delete.append(columnName)
        segment32delete.append(columnName)
      elif (segment15count > segment14count) and (segment15count > segment32count):
        print('Segment 15 wins')
        segment14delete.append(columnName)
        segment32delete.append(columnName)
      elif (segment32count > segment14count) and (segment32count > segment15count):
        print('Segment 32 wins')
        segment14delete.append(columnName)
        segment15delete.append(columnName)
      else: # there is a tie, the tokens is destroyed
        segment14delete.append(columnName)
        segment15delete.append(columnName)
        segment32delete.append(columnName)
        elementsDeleted.append(columnName)
print('Total terms ', i)
```
