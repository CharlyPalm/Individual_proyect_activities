```python
#Assuuming 'df' is your DataDrame
quality_counts = df['quality'].value_counts()
```

```python
#Using Matplotlib to create a count plot
plt.figure(figsize=(8,6))
plt.bar(quality_counts.index,quality_counts,color="pink")
plt.title("Count Plot of Quality")
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

```python
#Assuuming 'df' is your DataDrame
alcohol_counts = df['alcohol'].value_counts()
alcohol_counts
```

```python
#Using Matplotlib to create a count plot
plt.figure(figsize=(8,6))
plt.bar(alcohol_counts.index,alcohol_counts,color="blue")
plt.title("Count Plot of Quality")
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

```python
#Set Seaborn styke
sns.set_style("darkgrid")
```

```python
# Identify numerical columns
numerical_columns = df.select_dtypes(include=["int64","float64"]).columns
```

```python
#Plot distribution of each numerical feature
plt.figure(figsize=(14, len(numerical_columns) *3))
for idx, feature in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 3, idx)
    sns.histplot(df[feature],kde=True)
    plt.title(f"{feature} | Skewness: {round(df[feature].skew(),2)}")

#Adjust layout and show plots
plt.tight_layout()
plt.show()
```

```python
#Assuming 'df' is your DataFrame
plt.figure(figsize=(10,8))

#Using Seaborn to create a warm plot
sns.swarmplot(x="quality",y="alcohol",data=df,palette='viridis')

plt.title('Swarm Plot for Quality and Alcohol')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.show()
```

```python
# Set the color palette
sns.set_palette("Pastel1")

#Assuming'df' is your DataFrame
plt.figure(figsize=(10,6))

#Using Seaborn to create a pair plot with the specifies color palette
sns.pairplot(df)

plt.suptitle('Pair Plot for DataFrame')
plt.show()

plt.savefig('../data/processed')
```

```python
#Assuming 'df' is your DataFrame
df['quality'] = df['quality'].astype(str) #Convert 'quality' to categorical

plt.figure(figsize=(10,8))

#Using Seaborn to create a violin plot
sns.violinplot(x="quality",y="alcohol",data=df,palette={
    '3':'lightcoral','4':'lightblue','5':'lightgreen','6':'gold','7':'lightskyblue','8':'lightpink'}, alpha=0.7)
plt.title('Violin Plot for Quality and Alcohol')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.show()
```

```python
#plotting box plot between quality and alcohol
sns.boxplot(x='quality',y='alcohol',data=df)
```
