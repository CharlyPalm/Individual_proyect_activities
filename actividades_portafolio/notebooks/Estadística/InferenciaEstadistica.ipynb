# Inferencia Estadística

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('../data/raw/Inferencia Estadística.xlsx')
df
```

## Calculando el Nivel de Confianza de Media

```python
# Calculando el Interlvalo de Confianza al 95% para la media de la población
# Usando la fórmula de la media muestral
n = df['Consumo ajustado'].count() #Tamaño de la muestra
print(n)

x_bar = df['Consumo ajustado'].mean() #Promedio
print(x_bar)

s = df['Consumo ajustado'].std() #Desviación estándar
print(s)

n_conf = 0.95 #Nivel de confianza

# Calculo de z-critical value
import scipy.stats as stats
z_critical = stats.norm.ppf((1+n_conf)/2)

margin_error = z_critical * (s/np.sqrt(n))

# Calculo del margen de error
limit_sup = x_bar + margin_error
limit_inf = x_bar - margin_error

print('El intervalo de confianza al',n_conf*100,'% para la media de la población es: (', limit_inf, ',', limit_sup, ')')
```

## Calculando el nivel de cofianza con la media desconocida

```python
n = df['Consumo ajustado'].count() #Tamaño de la muestra
print(n)

x_bar = df['Consumo ajustado'].mean() #Promedio
print(x_bar)

s = df['Consumo ajustado'].std() #Desviación estándar
print(s)

n_conf = 0.90 #Nivel de confianza

t_critical = stats.t.ppf((1+n_conf)/2, n-1)

margin_error = t_critical * (s/np.sqrt(n))

# Calculo del margen de error
limit_sup = x_bar + margin_error
limit_inf = x_bar - margin_error

# W o diferencia entre los dos intervalos de confianza
w = limit_sup - limit_inf

print('El intervalo de confianza al', n_conf*100, '% para la media de la población es: (', limit_inf, ',', limit_sup, ')')
print('El margen de error es:', margin_error)
print("La diferencia entre los dos intervalos de confianza es:", w)

```

```python
# Calculando el nivel de muestra dado un margen de error y el porcentaje de confianza

n_conf = 0.97 #Nivel de confianza
w = 1.1 #Margen de error

z_critical = stats.norm.ppf((1+n_conf)/2)

n = ((2*z_critical * s) / w) ** 2
print(round(n,2))

print('El tamaño de la muestra necesario para un margen de error de', w, 'y un nivel de confianza de', n_conf*100, '% es:', n, ' \nPor lo tanto se necesitan:', np.ceil(n), 'muestras')

```

### Cálculo del intervalo de confianza para el 90% de la media de la población 

```python
n = df['Peso en Libras'].count() #Tamaño de la muestra
print('Hay', n, 'muestras')

x_bar = df['Peso en Libras'].mean() #Promedio
print('El promedio es:', x_bar)

s = df['Peso en Libras'].std() #Desviación estándar
print('La desviación estándar es:', s)

n_conf = 0.90 #Nivel de confianza

# Calculo de z-critical value
import scipy.stats as stats
z_critical = stats.norm.ppf((1+n_conf)/2)

margin_error = z_critical * (s/np.sqrt(n))

# Calculo del margen de error
limit_sup = x_bar + margin_error
limit_inf = x_bar - margin_error

diferencia = limit_sup - limit_inf

print('El intervalo de confianza al',n_conf*100,'% para la media de la población es: (', limit_inf, ',', limit_sup, ')')
print('La diferencia entre los dos intervalos de confianza es:', diferencia)
```

```python
n_conf = 0.98 #Nivel de confianza

chi_lower = stats.chi2.ppf((1-n_conf)/2, n-1)
chi_upper = stats.chi2.ppf((1+n_conf)/2, n-1)

lower_var = (n-1) * (s**2) / chi_upper
upper_var = (n-1) * (s**2) / chi_lower

lower_std = np.sqrt(lower_var)
upper_std = np.sqrt(upper_var)

diferencia = upper_std - lower_std

print('El intervalo de confianza al', n_conf*100, '% para la varianza de la población es: (', lower_std, ',', upper_std, ')')
print('La diferencia entre los dos intervalos de confianza es:', diferencia)
```

```python
# Calculando el nivel de muestra dado un margen de error y el porcentaje de confianza

n_conf = 0.95 #Nivel de confianza
w = 1.25 #Margen de error

z_critical = stats.norm.ppf((1+n_conf)/2)

n = ((2*z_critical * s) / w) ** 2
print(round(n,2))

print('El tamaño de la muestra necesario para un margen de error de', w, 'y un nivel de confianza de', n_conf*100, '% es:', n, ' \nPor lo tanto se necesitan:', np.ceil(n), 'muestras')
```
