# Tratamiento de Datos / Transformación de datos

```python
# Estadísticas descriptivas de las columnas 'GM', 'DR' y 'VS'
df[['GM', 'DR', 'VS']].describe()
```

```python
print("El coeficiente de variación es: \n", df[['GM', 'DR', 'VS']].std()/df[['GM', 'DR', 'VS']].mean())
print("\n Rango de las columnas: \n", df[['GM', 'DR', 'VS']].max()-df[['GM', 'DR', 'VS']].min())
```

```python
# Cálculo de la distancia Euclidiana entre los registros 1 y 3
import numpy as np

distancia = np.sqrt((df.loc[0, 'GM'] - df.loc[2, 'GM'])**2 + (df.loc[0, 'DR'] - df.loc[2, 'DR'])**2 + (df.loc[0, 'VS'] - df.loc[2, 'VS'])**2)
print("La distancia euclidiana entre los registros 1 y 3 es: ", distancia)
```

## Aplicando métodos de Estandarización 

### Método Max-Min (Feature Scaling o MinMax Scaler)
### Z = (x - min(x)) / (max(x)-min(x))

```python
# Método Max-Min para normalizar los datos
df_normMM = df
df_normMM
```

```python
df_normMM = (df - df.min())/(df.max() - df.min())
df_normMM
```

```python
df_normMM.describe()
```

```python
# Boxplot de las columnas normalizadas (Feature Scaling o MinMax Scaler)
plt.figure(figsize=(10, 6))
plt.title('Boxplot de las columnas GM, DR y VS')
plt.boxplot([df_normMM['GM'], df_normMM['DR'], df_normMM['VS']])
plt.ylabel('Valor')
plt.grid()
plt.show()
```

### Método Escalado Estándar 
### Z = (x - μ) / σ

```python
# Método Escalado Estándar para normalizar los datos
df_normEE = df
df_normEE
```

```python
df_normEE = (df - df.mean())/df.std()
df_normEE
```

```python
df_normEE.describe()
```

```python
# Boxplot de las columnas normalizadas (Método Escalado Estándar)
plt.figure(figsize=(10, 6))
plt.title('Boxplot de las columnas GM, DR y VS')
plt.boxplot([df_normEE['GM'], df_normEE['DR'], df_normEE['VS']])
plt.ylabel('Valor')
plt.grid()
plt.show()
```

### Método Escalado Estándar
### Z = x / σ 

```python
# Método Escalado Estándar para normalizar los datos
df_normEE2 = df
df_normEE2
```

```python
df_normEE2 = (df/df.std())
df_normEE2
```

```python
df_normEE2.describe()
```

```python
# Boxplot de las columnas normalizadas (Método Escalado Estándar)
plt.figure(figsize=(10, 6))
plt.title('Boxplot de las columnas GM, DR y VS')
plt.boxplot([df_normEE2['GM'], df_normEE2['DR'], df_normEE2['VS']])
plt.ylabel('Valor')
plt.grid()
plt.show()
```

```python
df2 = pd.read_excel('../data/raw/TransformaciondeDatos.xlsx', 'colinealidad interaccion')
df2
```

```python
# Generando plot de regresión y calculando la ecuación de regresión
sns.pairplot(df2, kind='reg')
```

```python
# Generando plot de regresión y calculando la ecuación de regresión
sns.lmplot(x='Resistencia', y='Molde', data=df2)
```

```python
#Ecuación de regresión del df2 siendo 'Resistencia' la variable dependiente, 'Termperatura' y 'Molde' la variable independiente
from sklearn.linear_model import LinearRegression

X = df2[['Temperatura', 'Molde']]
y = df2['Resistencia']

reg = LinearRegression().fit(X, y)
print("Coeficiente de determinación R^2: ", reg.score(X, y))
print("Intercepto: ", reg.intercept_)
print("Coeficientes: ", reg.coef_)
print("Ecuación de regresión: ", "Resistencia = ", reg.intercept_, "+", reg.coef_[0], "*Temperatura +", reg.coef_[1], "*Molde")
```
