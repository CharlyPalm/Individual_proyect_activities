### c) Estime, con una confianza de 94%, ¿cuál sería la resistencia promedio de los envases?

```python
# Calculando el nivel de confianza del 94% para la media de la resistencia de la botella
from scipy.stats import norm
import numpy as np

n = len(df)
media = df.mean()
desviacion_estandar = df.std()

n_confianza = 0.94

#Calculo de z-critical value
z_critical = norm.ppf((1+n_confianza)/2)

#Calculo de margen de error
margen_error = z_critical * (desviacion_estandar / np.sqrt(n))

# Calculo de margen de error
limite_inferior = media - margen_error.iloc[0]
limite_superior = media + margen_error.iloc[0]

dif = limite_superior.iloc[0] - limite_inferior.iloc[0]

print("El margen de error es:", margen_error.iloc[0].round(4))
print("La diferencia entre los límites es:", dif)
print(f"El intervalo de confianza para la media de {n_confianza*100}% la resistencia de la botella es: ({limite_inferior.iloc[0].round(4)} - {limite_superior.iloc[0].round(4)})")
```

El promedio de las botellas está entre 26.88kg y 27.60kg, con un nivel de confianza del 94%

### D.Antes del estudio se suponía que la resistencia promedio era de 25kg. Dada la evidencia de los datos, ¿tal supuesto es correcto? ¿Qué tipo de prueba estadística se debe realizar? Planteé las hipótesis correspondientes y concluya adecuadamente.

```python
# Prueba de Hipótesis

# Hipótesis nula: La resistencia promedio de la botella es de 25
# Hipótesis alternativa: La resistencia promedio de la botella no es de 25
import scipy.stats as stats
import numpy as np

#Libreria para calclular p-value
from scipy.stats import t


media = df.iloc[1:].mean()
media = media.iloc[0]

mu0 = 25

s = df.std()
s = s.iloc[0]

n = len(df)

alpha = 0.05

# Calculando el valor crítico z
z_critical = (media - mu0) / (s / np.sqrt(n))

# Calculando el valor p
p = 2 * (1 - stats.t.cdf(abs(z_critical), n-1))

print("El valor crítico z es:", z_critical)
print("El valor p es:", p)

if p < alpha:
    print("Rechazamos la hipótesis nula")
else:
    print("No rechazamos la hipótesis nula")
```

En este caso al ser mucho menor el P_value que alpha establecido, podemos decir que rechazamos la hipótesis nula, es decir, los datos no tenían un promedio de 25kg de resistencia.

### e) Con los datos anteriores estime, con una confianza del 98%, ¿cuál es la desviación estándar poblacional (del proceso)?

#### Utilizando chi-cuadrada

```python
# Calculando la desviación estándar de la resistencia de la botella dado un nivel de confianza del 98%
import scipy.stats as stats
import numpy as np

n = len(df) # Tamaño de la muestra
s = df.std() # Desviación estándar de la muestra
s = s.iloc[0] # Valor de la desviación estándar
n_confianza = 0.98 # Nivel de confianza

# Chi-Squared Critical values
chi2_lower = stats.chi2.ppf((1 - n_confianza) / 2, n-1)
chi2_upper = stats.chi2.ppf((1 + n_confianza) / 2, n-1)

lower_var = (n-1) * s**2 / chi2_upper
upper_var = (n-1) * s**2 / chi2_lower

lower_std = np.sqrt(lower_var)
upper_std = np.sqrt(upper_var)

print(f"El intervalo de confianza para la desviación estándar de la resistencia de la botella es: ({lower_std.round(4)} - {upper_std.round(4)}) para un nivel de confianza del {n_confianza*100}%")
```

### b. ¿La temperatura promedio más confortable es igual para hombre que para mujeres? ¿Qué tipo de prueba estadística se debe realizar? Planteé las hipótesis correspondientes y concluya adecuadamente.

Se debe de realizar una prueba de hipotesos de dos muestras, en este caso. Prueba Z

```python
# Pruebas de hipotesis de dos muestras
mu1 = 70
mu2 = 70
sigma1 = df.iloc[0:11, 0].std() # Desviación estándar para Mujeres
sigma2 = df.iloc[0:11, 1].std() # Desviación estándar para Hombres


n1 = len(df.iloc[0:11, 0]) # Tamaño de la muestra para Mujeres
n2 = len(df.iloc[0:11, 1]) # Tamaño de la muestra para Hombres

mean1 = df.iloc[0:11, 0].mean() # Media para Mujeres
mean2 = df.iloc[0:11, 1].mean() # Media para Hombres

print(f"Media para Mujeres: {round(mean1, 2)}% Desviación estándar para Mujeres: {round(sigma1, 5)}")
print(f"Media para Hombres: {round(mean2, 2)}% Desviación estándar para Hombres: {round(sigma2, 5)}")

z_score = (mean1 - mean2) / np.sqrt(sigma1**2/n1 + sigma2**2/n2)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print(f"Z-Score: {round(z_score, 5)}")
print(f"P-Value: {round(p_value, 5)}")

alpha = 0.05
if p_value < alpha:
    print("Rechazamos la hipótesis nula")
else:
    print("No rechazamos la hipótesis nula")
```

Podemos decir que las temperaturas promedio entre hombre y mujer son diferentes debido a que se tiene la evidencia para rechazar la hipótesis nula.

### c) ¿Los datos poseen la misma variabilidad? ¿Qué tipo de prueba estadística se debe realizar? Planteé las hipótesis correspondientes y concluya adecuadamente.

La hipótesis Nula es que las varianzas son las mismas

```python
# Prueba de hipótesis de dos muestras con varianzas desconocidas
mu1 = 70
mu2 = 70

var1 = df.iloc[0:11, 0].var() # Varianza para Mujeres
var2 = df.iloc[0:11, 1].var() # Varianza para Hombres

f_statistic = var1 / var2

df1 = len(df.iloc[0:11, 0]) - 1
df2 = len(df.iloc[0:11, 1]) - 1

p_value = 2 * min(stats.f.cdf(f_statistic, df1, df2), 1 - stats.f.cdf(f_statistic, df1, df2))

print(f"F-Statistic: {round(f_statistic, 5)}")
print(f"P-Value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Rechazamos la hipótesis nula")
else:
    print("No rechazamos la hipótesis nula")

```

Para encontrar una variabilidad se puede utilizar una Prueba F, la cuál nos ayudará a encontrar la diferencia entre las varianzas y también generar una respuesta a la hipótesis nula, la cuál es que la no tienen variabilidad.
