```python
# Calculando los p-values

from scipy.stats import ttest_ind

# Calculando con un nivel de significancia de 0.05
n_confianza = 0.95


#Columna A vs B
AB = ttest_ind(df['A'], df['B'], equal_var=False).pvalue
print("El p-value de la comparación entre las columnas A y B es: ", AB)

#Columna A vs C
AC = ttest_ind(df['A'], df['C'], equal_var=False).pvalue
print("El p-value de la comparación entre las columnas A y C es: ", AC)

#Columna A vs D
AD = ttest_ind(df['A'], df['D'], equal_var=False).pvalue
print("El p-value de la comparación entre las columnas A y D es: ", AD)

#Columna B vs C
BC = ttest_ind(df['B'], df['C'], equal_var=False).pvalue
print("El p-value de la comparación entre las columnas B y C es: ", BC)

#Columna B vs D
BD = ttest_ind(df['B'], df['D'], equal_var=False).pvalue
print("El p-value de la comparación entre las columnas B y D es: ", BD)

#Columna C vs D
CD = ttest_ind(df['C'], df['D'], equal_var=False).pvalue
print("El p-value de la comparación entre las columnas C y D es: ", CD)
```
